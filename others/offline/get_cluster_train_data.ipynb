```python
import jieba
```

```python
from gensim.models import KeyedVectors
import numpy as np
# from utils import write_a_log
from config import WORD2VER_MODEL_PATH
word2vec_model = KeyedVectors.load(WORD2VER_MODEL_PATH, mmap='r')
```

```python
STOPWORDS = load_stopwords(DEFAULT_STOPWORDS_PATH) 
```

```python
df_available = df[(df.speech.str.len()>5)&(df.speech.str.len()<150)]
```

```python
word2vec_model.vector_size
```

```python
def geneVector(speech):
    words =  [w for w in jieba.cut(speech) if w not in STOPWORDS]
    vectors = [word2vec_model[w] for w in words if w in word2vec_model]
    ret = np.average(vectors,axis=0).reshape(100) if len(vectors)>0 else np.zeros(word2vec_model.vector_size).reshape(100)
    return ret
```

```python
df_available.drop('vector',inplace=True,axis=1)
```

```python
def generate_ner(doc):
    box = []
    if doc.id%1000==0:
        print(doc.id,end='|')
    try:
        clean_doc = doc.content.replace('\\n','').replace('\u3000','')

        words = LTPM.split_words(clean_doc)
        postags = LTPM.get_sentence_pos(words)
        ner = LTPM.NER(words,postags)
        ni_or_nh = [(w,tag) for w,tag in zip(words,ner) if tag.endswith('Nh') or tag.endswith('Ni')]
        curr = ''
        for w,tag in ni_or_nh:
            if tag.startswith('S'):
                box.append(w)
            if tag.startswith('B'):
                curr = ''
                curr += w
            if tag.startswith('I'):
                curr += w
            if tag.startswith('E'):
                curr += w
                box.append(curr)
    except Exception as e:
        print('err',doc.id,end=',')
    return box

df_raw['NER'] = df_raw.apply(generate_ner,axis=1)
```
